```python
# Inventors cleaning
#inventionKey = inventors['inventionsKey'].value_counts().keys().tolist()
#numberofpeople = inventors['inventionsKey'].value_counts().tolist()
#inventors=pd.DataFrame(list(zip(inventionKey, numberofpeople)),columns=['inventionsKey','numberofpeople'])
inventors["department"] = inventors["dept"].map(str) + inventors["company"]
inventors.drop("dept",axis=1,inplace=True)
inventors.drop("company",axis=1,inplace=True)
inventors.drop("leadInventor",axis=1,inplace=True)
inventors=inventors.groupby('inventionsKey').agg('nunique')
inventors['inventionsKey'] = inventors.index
```

```python
#numeric coding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
var_to_encode = ['type','revenue','countryCode','status']
for col in var_to_encode:
    inv_new[col] = le.fit_transform(inv_new[col].astype(str))
```

```python
inv_new.fillna(value=-1,inplace=True)
```

```python

var_to_normalize = ['peopleKey','department','fundingAmount','type','countryCode','status']
for col in var_to_normalize:
    print(1)
    #mean = np.mean(inv_new[col])
    max = np.max(inv_new[col])
    min = np.min(inv_new[col])
    inv_new[col]= inv_new[col].apply(lambda x: (x - min)/(max-min))
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(inv_new, test_size=0.2)
```
